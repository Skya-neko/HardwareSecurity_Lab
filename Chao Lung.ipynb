```python
target = hw_list[AES_Sbox[k^p]]
```

```python
SNRplot = []
for x in range(0,16):
    listHDGP = {0:[], 1:[], 2:[], 3:[], 4:[], 5:[], 6:[], 7:[], 8:[]}
    for i in range(len(target[x])):
        for j in range(len(target)): 
                if target[j][x] == i:
                    listHDGP[i].append(t[j])  
    meanlist = []
    varlist = []
    for i in range(0,9):
        meanlist.append(np.mean(listHDGP[i],axis = 0))
        varlist.append(np.var(listHDGP[i],axis = 0))

    numerator = np.var(meanlist,axis = 0)  
    denominator = np.mean(varlist,axis = 0)

    SNR2 = np.array(numerator/denominator)
    SNRplot.append(SNR2)
```
