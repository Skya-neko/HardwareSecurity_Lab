```python
#k,p做XOR運算後，放入AES查表，再將值放入hw_list查表
hd_list = hw_list[AES_Sbox[k^p]]   #50000*16



#做Template attack必須先蒐集POI及Template
TSNRplot = []   #randkey裡16個byte的SNR圖
TPOI_des = []   #randkey裡16個byte的POI位置，有16*10個
TTemplate = []  #randkey裡16個byte的Template(一個byte有9個不同HD的Template)(每一個Template由mean vector, covariance構成)

#16個bytes
for x in range(0,16):   
    
    #分群，0~9的HD
    listHDGP = [[] for i in range(0,9)]
    for i in range(len(hd_list[x])):          #16 
        for j in range(len(hd_list)):         #50000
            if hd_list[j][x] == i:
                listHDGP[i].append(t[j])  

    #分群權重
    listHDGP_w = [ len(listHDGP[i])/50000 for i in range(len(listHDGP))]
     
    #SNR之第一次計算分子分母
    meanlist = []
    varlist = []
    for i in range(0,9):
        meanlist.append(np.mean(listHDGP[i],axis = 0))
        varlist.append(np.var(listHDGP[i],axis = 0))
    
    #SNR之第二次計算分子分母-->做權重
    son = np.average(np.array(np.power(meanlist,2)), axis=0, weights=listHDGP_w) - np.power(np.average(np.array(meanlist), axis=0, weights=listHDGP_w),2)
    mother = np.average(np.array(varlist), axis=0, weights=listHDGP_w)

    #SNR之分子除以分母
    SNR = np.array(son/mother)
    TSNRplot.append(SNR)
    
    #取出10個POI
    POI_des = np.argsort(SNR)[::-1][:10]
    TPOI_des.append(POI_des)
    
    #Template Building (Cont.)
    Cov = []
    Mean = []
    for i in range(len(listHDGP)): #0,9
        byte15HDq = []
        for q in range(len(listHDGP[i])):
            byte15HDj = []
            for j in POI_des:
                byte15HDj.append(listHDGP[i][q][j])  #取在第i群的POI trace，每一群有q個，就有q*10，用q*10算出唯一一個cov
            byte15HDq.append(byte15HDj)
        Mean.append(np.mean(np.array(byte15HDq),axis = 0))
        Cov.append(np.cov(np.array(byte15HDq).T))  #轉置矩陣
        
    #Template Building    
    Template = []
    for i in range(0,9):
        Template.append([np.array(Mean[i]),np.array(Cov[i])])
    TTemplate.append(Template)
    
    
```
